```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
headerList = ['#', 'gold', 'silver', 'CTAB', 'I-2959','UV_Inten', 'time', 'cycle', 'num_peak','wl1', 'FWHM1', 'ISPR1', 'wl2', 'FWHM2','ISPR2','I400','note']
data = pd.read_csv(r"All Data cleaned_v2.csv", names=headerList)
data =  data.iloc[1:]
# data
# Data format
data_num = data.copy()
numeric = [ 'gold','silver', 'CTAB', 'I-2959','UV_Inten', 'time','cycle']
data_num[numeric] = data_num[numeric].apply(pd.to_numeric, errors='coerce')
data_num
```

```python
# data.shape
# data.describe()
# data.info()
data_num.note.value_counts()
```

```python
# Dealing with unwanted/missing data
clf_data1 = data_num.drop(data_num[(data_num['note'] == '0')].index)
clf_data2 = clf_data1.drop(clf_data1[(clf_data1['note'] == 'XR')].index)
data_clf = clf_data2.drop(clf_data2[(clf_data2['note'] == 'TBD')].index)
data_clf
# data_clf.note.value_counts()
```

```python
# create a 2nd dataset for different parameters
data2 = data_clf
data2['Power'] = data2['UV_Inten']*data2['time']
data2
```

```python
X = data_clf[[ 'gold','silver', 'CTAB', 'I-2959','UV_Inten', 'time','cycle']]
X2 = data2[[ 'gold','silver', 'CTAB', 'I-2959','Power','cycle']]
y = data_clf[['note']]
label_map = {"NT":3, 'R':2, 'CS':1, 'S':0}
train_labels = data_clf['note'].map(label_map)
classes = [
    'Alloy',
    'Core-shell',
    'AuNR',
    'Nanotetrapod'
    ]
y_dummy = pd.get_dummies(y['note'])
y_dummy
```

# Data normalization

```python
X_norm = X.copy()

# X_norm.iloc[:,0:7] = X_norm.iloc[:,0:7].apply(lambda x: (x/x.max()), axis=0)
X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.3)
```

# Random Forest feature importance

Use a Random Forest classifier to calculate the feature importance scores. Random Forest determines the importance of each feature by measuring the decrease in model performance (such as Gini impurity or entropy) when a feature is randomly permuted. Higher feature importance scores indicate greater importance. May work well with non-linear relationships

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
print(rf.score(X_train, y_train))
print(rf.score(X_test, y_test))
importance_scores = rf.feature_importances_
feature_importance_df = pd.DataFrame({'Parameter': X_norm.columns, 'Importance Score': importance_scores})
print(feature_importance_df)
feature_importance_df.sort_values(by='Importance Score', ascending=False, inplace=True)
selected_features_rf = feature_importance_df['Parameter'].tolist()
print("Features selected by Random Forest:", selected_features_rf)
```

```python
from sklearn.inspection import permutation_importance
pis = permutation_importance(rf, X_norm, y, n_repeats=10, random_state=0)['importances_mean']
pis_df = pd.DataFrame({'Parameter': X_norm.columns, 'Importance Score': pis})
print(pis_df)
```

# Recursive Feature Elimination

Use Recursive Feature Elimination (RFE) with cross-validation. RFE works by recursively removing features from the model and selecting the features that contribute most to the model's performance. By using cross-validation, RFE helps to avoid overfitting and selects the optimal subset of features.

```python
rfe_selector = RFE(estimator=RandomForestClassifier(), n_features_to_select=3)
rfe_selector.fit(X, y)
selected_features_rfe = X.columns[rfe_selector.support_].tolist()
print("Features selected by RFE:", selected_features_rfe)
```

# Model Building: Ensemble Trees Model

```python
# random Forest
clf0= RandomForestClassifier(n_estimators=100)
clf0.fit(X_train,Y_train)
std = np.std([tree.feature_importances_ for tree in clf0.estimators_], axis=0)
feature_importances = pd.Series(clf0.feature_importances_, index=X_train.columns)
print(feature_importances)
print('Score of RF model on test split\n',clf0.score(X_test,Y_test))
```

```python
#Gradient Boosing Tree
clf= GradientBoostingClassifier(n_estimators=100,random_state=1200)
clf.fit(X_train,Y_train)

feature_importances = pd.Series(clf.feature_importances_, index=X_train.columns)
print(feature_importances)
print('Score of Gradient Boosting model on test split\n',clf.score(X_test,Y_test))
```
