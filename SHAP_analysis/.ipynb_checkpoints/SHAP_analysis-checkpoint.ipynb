```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import plotly
np.random.seed(0)
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler   # Library for data preprocessing
from sklearn.decomposition import PCA              # Library for Principal Component Analysis
import shap                                        # Library for SHAP (SHapley Additive exPlanations)
from sklearn.svm import SVR                         # Support Vector Regression from scikit-learn
from sklearn.model_selection import cross_val_score, GridSearchCV  # Tools for model evaluation and tuning
from sklearn.model_selection import KFold           # K-Fold cross-validation
import xgboost as xgb                               # XGBoost library for gradient boosting
```

```python
# Defining column headers for the DataFrame
headerList = ['#', 'gold', 'silver', 'CTAB', 'I-2959','UV_Inten', 'time', 'cycle', 'num_peak','wl1', 'FWHM1', 'ISPR1', 'wl2', 'FWHM2','ISPR2','I400','note']

# Reading data from a CSV file with custom column names
data = pd.read_csv(r"NR_raw_data.csv", names=headerList)

# Removing the first row (assumed to be headers) to keep only the data
data = data.iloc[1:]

# Creating a copy of the DataFrame for numeric data conversion
data_num = data.copy()

# Specifying the columns to be converted to numeric
numeric = ['gold', 'silver', 'CTAB', 'I-2959', 'UV_Inten', 'time', 'cycle', 'wl1', 'FWHM1', 'ISPR1', 'wl2', 'FWHM2', 'ISPR2', 'I400']

# Converting specified columns to numeric, handling any errors
data_num[numeric] = data_num[numeric].apply(pd.to_numeric, errors='coerce')

# Displaying the DataFrame with numeric data
data_num
```

# Gold Nanorods

```python
# Creating a new DataFrame 'data_R' containing rows where 'note' is 'R'
data_R = data_num.drop(data_num[(data_num['note'] != 'R')].index)

# Adding a new column 'IR' calculated by dividing 'ISPR1' by 'I400'
data_R['IR'] = data_R['ISPR1'] / data_R['I400']

# Displaying the resulting DataFrame 'data_R'
data_R
```

```python
def trainModel(X, y):
    """
    Train a model using XGBoost with hyperparameter tuning.
    
    Parameters:
    - X: Features data for training.
    - y: Target labels for training.
    
    Returns:
    - clf: Trained model with the best hyperparameters.
    """
    
    # Hyperparameter grid for tuning
    params = {
        'max_depth': [3, 6, 10],                  # Maximum depth of trees
        'learning_rate': [0.01, 0.05, 0.1],       # Learning rate for boosting
        'n_estimators': [100, 500, 1000],         # Number of boosting rounds
        'colsample_bytree': [0.3, 0.7]            # Fraction of features used for each tree
    }
    
    # Creating an XGBoost classifier model
    model = xgb.XGBClassifier()
    
    # Initializing GridSearchCV for hyperparameter tuning
    clf = GridSearchCV(estimator=model, param_grid=params)
    
    # Fitting the classifier to the training data
    clf.fit(X, y)
    
    return clf
```

```python
def shap_analysis(X, y, model):
    """
    Perform SHAP analysis on the given model using the provided data.
    
    Parameters:
    - X: Features data for analysis.
    - y: Target labels for analysis.
    - model: Trained model for analysis.
    
    Returns:
    None
    """
    
    # Creating a KFold cross-validation object with 10 splits
    kf = KFold(n_splits=10, shuffle=True)
    
    # Printing the average R^2 score of cross-validated predictions
    print(target, cross_val_score(model, X, y, cv=kf, scoring='r2').mean())
    
    # Fitting the model to the entire dataset
    clf = model.fit(X, y)
    
    # Printing the R^2 score of the model (non-CV)
    print(target, 'non-CV', clf.score(X, y))
    
    # Predicting on the dataset and calculating residuals
    y_pred = model.predict(X=X)
    print(y_pred - y)
    
    # Creating a scatter plot of residuals
    plt.scatter(np.arange(len(y)), y_pred - y)
    plt.title(target)
    plt.xlabel('exp #')
    plt.ylabel('residual')
    plt.show()
    plt.clf()
    
    # Creating a SHAP explainer using the model's predictions
    explainer = shap.KernelExplainer(model.predict, X)
    
    # Calculating SHAP values
    shap_values = explainer.shap_values(X)
    
    # Creating a summary plot of SHAP values
    shap.summary_plot(shap_values, features=X, feature_names=features, title=target, show=False)
    
    # Saving the summary plot as an image file
    plt.title(target)
    plt.savefig(target + '_AuNR.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
# Defining the list of features and targets
features = ['gold', 'silver', 'CTAB', 'I-2959', 'UV_Inten', 'time', 'cycle']
targets = ['wl1', 'FWHM1', 'IR']

# Extracting the feature data
X = data[features]

# Looping through each target
for t in targets:
    # Extracting the target data
    y = data[t]
    
    # Training a model using the defined features and target
    model = trainModel(X, y)
    
    # Performing SHAP analysis on the trained model
    shap_analysis(X, y, model)
```

```python
# Importing necessary libraries
%matplotlib inline
features = ['gold', 'silver', 'CTAB', 'I-2959', 'UV_Inten', 'time', 'cycle']
targets = ['wl1', 'FWHM1', 'IR']

# Creating a StandardScaler instance
scaler = StandardScaler()

# Extracting raw feature data
X_raw = np.array(data_R.loc[:, features])

# Looping through each target
for target in targets:
    # Extracting raw target data
    y_raw = np.array(data_R.loc[:, [target]])
    
    # Fitting the scaler to raw feature data and transforming it
    scaler.fit(X_raw)
    X_scaled = scaler.transform(X_raw)
    
    # Fitting the scaler to raw target data, transforming it, and flattening
    scaler.fit(y_raw.reshape(-1, 1))
    y_scaled = scaler.transform(y_raw.reshape(-1, 1)).ravel()
    
    # Performing SHAP analysis on scaled feature and target data
    shap_analysis(X_scaled, y_scaled, features, target)
```
