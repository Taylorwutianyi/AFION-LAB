```python
import pandas as pd
import numpy as np
import plotly
np.random.seed(0)
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import shap
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.model_selection import KFold
import xgboost as xgb
```

```python
headerList = ['#', 'gold', 'silver', 'CTAB', 'I-2959','UV_Inten', 'time', 'cycle', 'num_peak','wl1', 'FWHM1', 'ISPR1', 'wl2', 'FWHM2','ISPR2','I400','note']
data = pd.read_csv(r"All Data cleaned_v2.csv", names=headerList)
data =  data.iloc[1:]
# data
# Data format
data_num = data.copy()
numeric = [ 'gold','silver', 'CTAB', 'I-2959','UV_Inten', 'time','cycle','wl1','FWHM1','ISPR1','wl2','FWHM2','ISPR2','I400']
data_num[numeric] = data_num[numeric].apply(pd.to_numeric, errors='coerce')
data_num
```

```python
data_num.note.value_counts()
```

# Gold Nanorods

```python
data_R = data_num.drop(data_num[(data_num['note'] != 'R')].index)
data_R['IR'] = data_R['ISPR1']/data_R['I400']
data_R
```

```python
def trainModel(X, y):
    params = {'max_depth': [3,6,10],
           'learning_rate': [0.01, 0.05, 0.1],
           'n_estimators': [100, 500, 1000],
           'colsample_bytree': [0.3, 0.7]}
    model = xgb.XGBClassifier()
    clf = GridSearchCV(estimator=model, 
                   param_grid=params)
    clf.fit(X,y)
    return clf
```

```python
def shap_analysis(X, y, model):
    kf = KFold(n_splits=10, shuffle=True)

    print (target, cross_val_score(model, X, y, cv=kf, scoring='r2').mean())

    clf = model.fit(X, y)
    print(target, 'non-CV', clf.score(X, y))
    y_pred = model.predict(X=X)
    print (y_pred - y)
    plt.scatter(np.arange(len(y)), y_pred - y)
    plt.title(target)
    plt.xlabel('exp #')
    plt.ylabel('residual')
    plt.show()
    plt.clf()
    explainer = shap.KernelExplainer(model.predict, X)
    shap_values = explainer.shap_values(X)
    shap.summary_plot(shap_values, features=X, feature_names=features, title=target,show=False)
    plt.title(target)
    plt.savefig(target+'_AuNR.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
features = ['gold', 'silver', 'CTAB', 'I-2959', 'UV_Inten', 'time', 'cycle']
targets = ['wl1', 'FWHM1','IR']
X = data[features]
for t in targets:
    y = data[t]
    model = trainModel(X,y)
    shap_analysis(X, y, model)
```

```python
%matplotlib inline
features = ['gold', 'silver', 'CTAB', 'I-2959', 'UV_Inten', 'time', 'cycle']
targets = ['wl1', 'FWHM1','IR']
scaler = StandardScaler()
X_raw = np.array(data_R.loc[:, features])


for target in targets:
    y_raw = np.array(data_R.loc[:, [target]])
    scaler.fit(X_raw)
    X_scaled = scaler.transform(X_raw)
    scaler.fit(y_raw.reshape(-1,1))
    y_scaled = scaler.transform(y_raw.reshape(-1,1)).ravel()
    # print (X, y)
    shap_analysis(X_scaled, y_scaled, features, target)
   
```
